```r
tri_recursion <- function(k) {
  if (k > 0) {
    result <- k + tri_recursion(k - 1)
    print(result)
  } else {
    result = 0
    return(result)
  }
}
tri_recursion(6)
```

```r
x <- c(151, 174, 138, 186, 128, 136, 179, 163, 152, 131)
y <- c(63, 81, 56, 91, 47, 57, 76, 72, 62, 48)

relation <- lm(y~x)
print(relation)
```

```r
print(summary(relation))
```

```r
#predict
a <- data.frame(x=c(170,180,190))
result <- predict(relation, a)
print(result)
```

```r
#visualize the linear regression
plot(y,x,abline(lm(x~y)),main="Height & Weight Regression",
    xlab="Weight in Kg", ylab="Height in cm", col="blue",
    cex=1.3, pch=16)
```

```r
#multiple regression
#this can be done with lm() funtion also
#formula ---- y = a + b1x1 + b2x2 +...bnxn
```

```r
input <- mtcars[,c("mpg","disp","hp","wt")]
print(head(input))
```

```r
model <- lm(mpg~disp+hp+wt, data=input)
print(model)
```

```r
a <- coef(model)[1]
Xdisp <- coef(model)[2]
Xhp <- coef(model)[3]
Xwt <- coef(model)[4]
print(a)
print(Xdisp)
print(Xhp)
print(Xwt)
```

```r
library(tensorflow)
library(keras)
#The CIFAR10 dataset contains 60,000 color images in 10 classes, 
#with 6,000 images in each class. The dataset is divided into 
#50,000 training images and 10,000 testing images.
X_train <- array(rnorm(100 * 28 * 28), dim = c(100, 28, 28, 1))
Y_train <- to_categorical(matrix(sample(0:2, 100, TRUE), ncol = 1), 3)
```

```r
model <- keras_model_sequential()
model %>%
  layer_conv_2d(filters = 32, kernel_size = c(3,3), activation = "relu", 
                input_shape = c(28, 28, 1)) %>% 
  layer_max_pooling_2d(pool_size = c(2,2)) %>% 
  layer_conv_2d(filters = 64, kernel_size = c(3,3), activation = "relu") %>% 
  layer_flatten() %>%
  layer_dense(units = 64, activation = "relu") %>% 
  layer_dropout(0.25) %>%
  layer_dense(units = 3, activation = "softmax")
```

```r
summary(model)
```

```r
model %>% compile(optimizer="adam", loss="categorical_crossentropy",metrics="accuracy")
history <- model %>% fit(x=X_train, y=Y_train,
                        epochs=100,
                        verbose=2)
```

```r
plot(history)
```
